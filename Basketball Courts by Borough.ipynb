```python
import pandas as pd
```

```python
b_ball=pd.read_json('DPR_Basketball_001.json')
```

```python
b_ball['Borough']='Default Value'
#rowIndex=b_ball.index[0]
#a=b_ball.loc[rowIndex,'Borough']='Bronx'
```

```python
b_ball
```

```python
for index,row in b_ball.iterrows():
    if 'X'in row['Prop_ID']:
        b_ball.at[index,'Borough']='Bronx'
    elif 'B' in row['Prop_ID']:
        b_ball.at[index,'Borough'] ='Brooklyn'
    elif 'M' in row['Prop_ID']:
        b_ball.at[index,'Borough']='Manhattan'
    elif 'Q' in row['Prop_ID']:
        b_ball.at[index,'Borough']='Queens'
    else:
        b_ball.at[index,'Borough']='Staten Island'
b_ball
```

```python
borough_breakdown=b_ball[['Borough']]
```

```python
bb_down=borough_breakdown['Borough'].value_counts().to_frame().reset_index()
```
